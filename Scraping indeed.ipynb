```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

df_M = pd.DataFrame({'review_title': [],'review':[],'author':[],'rating':[],'pros':[]})
```

```python
#for Maersk 
for i in range(0,2700,20):
     url = (f'https://www.indeed.com/cmp/Maersk/reviews?fcountry=ALL&start={i}&lang=en')
     header = {"User-Agent":"Mozilla/5.0 Gecko/20100101 Firefox/33.0 GoogleChrome/10.0"}
     page = requests.get(url,headers = header)
     soup = BeautifulSoup(page.content, 'lxml')
     results = soup.find("div",'cmp-ReviewsList')
     elems = results.find_all(class_='css-5cqmw8 eu4oa1w0')
     for elem in elems:
         title = elem.find(attrs = {'class':'css-wn2j4x e1tiznh50'})
         review = elem.find(attrs = {'class': 'css-rr5fiy eu4oa1w0'})
         author = elem.find(attrs = {'class':'css-8a5o2x e1wnkr790'})
         rating = elem.find(attrs = {'class':'css-1h3aion eu4oa1w0'})
         pros = elem.find_all(class_='css-1z0411s e1wnkr790')
         df_M = df_M.append({'review_title': title.text,'review': review.text,'author': author.text,'rating': rating.text,'pros':pros}, ignore_index=True)
        
        
        
        
        
        
```
