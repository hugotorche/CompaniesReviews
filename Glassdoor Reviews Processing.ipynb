## **Bloomberg Data**

```python
# Load Data
bb_carlsberg_social = pd.read_excel('/content/Carlsberg Social Governance Data.xlsx', 'Social')
bb_carlsberg_social.head()
```

```python
# Set headers

bb_carlsberg_social = bb_carlsberg_social.iloc[2:,:]
bb_carlsberg_social.columns = bb_carlsberg_social.iloc[0,:]
bb_carlsberg_social.reset_index(inplace=True)
bb_carlsberg_social = bb_carlsberg_social.iloc[:,1:]


bb_carlsberg_social.head()
```

```python
# Set rows and columns

bb_carlsberg_social = bb_carlsberg_social.iloc[2:,1:]
bb_carlsberg_social.reset_index(inplace=True)

bb_carlsberg_social = bb_carlsberg_social.iloc[:,1:]

bb_carlsberg_social.head()
```

```python
# Transpose data

bb_carlsberg_social = bb_carlsberg_social.T
bb_carlsberg_social.columns = bb_carlsberg_social.iloc[0,:]
bb_carlsberg_social = bb_carlsberg_social.iloc[1:,:]
bb_carlsberg_social.head()
```

```python
# drop NaN columns 

bb_carlsberg_social.dropna(axis=1, how='all', inplace=True)
bb_carlsberg_social.head()
```

```python
# Replace -- and nan values

bb_carlsberg_social.replace('—','',inplace=True)
bb_carlsberg_social.fillna('', inplace=True)
bb_carlsberg_social.head()
```

```python
# Reset index
bb_carlsberg_social.reset_index(inplace=True)
bb_carlsberg_social = bb_carlsberg_social.iloc[:, 0:95]
bb_carlsberg_social.head()
```

```python
# Rename columns
bb_carlsberg_social.rename({2: "Year"}, axis='columns', inplace=True)
bb_carlsberg_social.rename(str.lower, axis='columns', inplace=True)

bb_carlsberg_social.head()
```

```python
# Set year column
bb_carlsberg_social['year'] = bb_carlsberg_social['year'].map(lambda x: x.lstrip('FY '))
bb_carlsberg_social.head()
```
